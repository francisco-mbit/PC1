EXPLICACIÓN DE VISUALIZACIÓN:

Cada boxplot nos permite ver:

¿Cómo interpretar la gráfica?

- La caja central representa el 50% de los datos (Q1 a Q3).
- La línea dentro de la caja es la mediana (valor central).
- Los bigotes representan el rango típico de valores dentro de 1.5 * IQR.
- Los puntos fuera de los bigotes son valores atípicos (outliers).

MÉTODO IQR:

El método del Rango Intercuartil (IQR) es una técnica estadística ampliamente usada para detectar y eliminar valores atípicos (outliers) en los datos. Se basa en los cuartiles y es útil porque no asume que los datos siguen una distribución normal, a diferencia de otros métodos como la desviación estándar.

¿Qué son los cuartiles?

Q1 (Primer cuartil, 25%): El valor debajo del cual se encuentra el 25% de los datos.
Q3 (Tercer cuartil, 75%): El valor debajo del cual se encuentra el 75% de los datos.
IQR (Rango Intercuartil): La diferencia entre Q3 - Q1, que nos da el rango donde se encuentra el 50% central de los datos.

¿Por qué 1.5 * IQR?

Es una regla empírica que detecta valores extremos.
Un valor menor que Q1 - 1.5 * IQR o mayor que Q3 + 1.5 * IQR se considera outlier.
Se elige 1.5 porque captura la mayoría de los valores típicos sin eliminar datos importantes.

Usamos después una máscara booleana para identificar filas con valores fuera de los límites permitidos.
Eliminamos cualquier fila donde al menos una variable numérica sea un outlier.

CONCLUSIÓN:

- El método IQR nos ayuda a detectar valores extremos sin asumir que los datos son normales.
- Usamos 1.5 * IQR porque es una regla estándar que mantiene la mayoría de los datos sin eliminar información útil.
- Aplicamos una máscara booleana para eliminar las filas que contienen outliers en cualquier columna numérica.
- Mostramos cuántos registros fueron eliminados para asegurarnos de que no estamos perdiendo demasiados datos.

```python
import numpy as np

# Calcular cuartiles y rango intercuartil
Q1 = df_combined[numeric_cols].quantile(0.25)
Q3 = df_combined[numeric_cols].quantile(0.75)
IQR = Q3 - Q1

# Definir límites para valores atípicos
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar valores dentro del rango permitido
df_filtered = df_combined[~((df_combined[numeric_cols] < lower_bound) | (df_combined[numeric_cols] > upper_bound)).any(axis=1)]

# Mostrar cuántos registros eliminamos
print(f"Registros antes de filtrar: {df_combined.shape[0]}")
print(f"Registros después de filtrar: {df_filtered.shape[0]}")
```

Importancia de esta limpieza:

- Evita que valores extremos distorsionen los análisis y modelos predictivos.

¿Por qué almacenar los datos en SQLite?

Persistencia de datos:

- Guardar en un archivo CSV está bien para análisis rápidos, pero una base de datos es más segura y estructurada.
- Con SQLite, los datos quedan guardados en un archivo .db que puedes consultar en cualquier momento.

Acceso eficiente:

- Si en el futuro quieres hacer consultas específicas sin cargar todo el dataset en memoria, SQLite lo permite con SELECT, WHERE, etc.
- Por ejemplo, puedes consultar solo los vinos con calidad mayor a 7 sin abrir todo el CSV.

Escalabilidad:

- Si decides hacer Machine Learning o Dashboards, tener los datos en una base de datos permite integrarlos con otras herramientas.
- SQLite se puede usar en aplicaciones web, análisis de datos y reporting.

# 5. REALIZAR CONSULTAS EN SQLITE

Basándote en los datos y las columnas del dataset, realiza las siguientes consultas:

Consulta 1: ¿Cuál es el promedio de calidad (`quality`) por tipo de vino (`type`)?

Consulta 2: ¿Cuántos vinos tienen un nivel de alcohol superior a 10.5, agrupados por tipo?

Consulta 3: Obtén el conteo de vinos por nivel de acidez (`fixed acidity`) agrupados en rangos (por ejemplo, de 0-5, 5-10, 10-15).

```python
# Conectar a la base de datos
conn = sqlite3.connect("wine_quality.db")

# Verificar qué tablas existen en la base de datos
tables = pd.read_sql("SELECT name FROM sqlite_master WHERE type='table';", conn)
print("Tablas en la base de datos:")
print(tables)

conn.close()
```

```python
# Conectar a la base de datos SQLite
conn = sqlite3.connect("wine_quality.db")
```

```python
# Consulta 1: Promedio de calidad por tipo de vino

query1 = """
SELECT type, AVG(quality) AS avg_quality
FROM wine_quality
GROUP BY type;
"""
df_query1 = pd.read_sql(query1, conn)
print("\n Promedio de Calidad por Tipo de Vino")
print(df_query1)
```

```python
# Consulta 2: Conteo de vinos con alcohol > 10.5, agrupados por tipo

query2 = """
SELECT type, COUNT(*) AS count_wines
FROM wine_quality
WHERE alcohol > 10.5
GROUP BY type;
"""
df_query2 = pd.read_sql(query2, conn)
print("\n Conteo de Vinos con Alcohol > 10.5")
print(df_query2)
```

```python
# Consulta 3: Conteo de vinos por nivel de acidez en rangos (0-5, 5-10, 10-15)

query3 = """
SELECT 
    CASE 
        WHEN "fixed acidity" BETWEEN 0 AND 5 THEN '0-5'
        WHEN "fixed acidity" BETWEEN 5 AND 10 THEN '5-10'
        WHEN "fixed acidity" BETWEEN 10 AND 15 THEN '10-15'
        ELSE '15+' 
    END AS acidity_range,
    COUNT(*) AS count_wines
FROM wine_quality
GROUP BY acidity_range;
"""
df_query3 = pd.read_sql(query3, conn)
print("\n Conteo de Vinos por Nivel de Acidez")
print(df_query3)
```

Vemos que en acidity_range no nos aparece 10-15 vamos a comprobar los valores máximos y mínimos de esa columna, si el valor máximo es menor a 10, entonces la categoría 10-15 nunca será usada porque no hay registros en ese rango como parece ser este caso.

```python
# Comprobamos los máximos y mínimos de fixed acidity

query_check = """
SELECT MIN("fixed acidity") AS min_acidity, 
       MAX("fixed acidity") AS max_acidity
FROM wine_quality;
"""
df_check = pd.read_sql(query_check, conn)
print("\n Rango de valores de fixed acidity en la base de datos")
print(df_check)
```

De cara a una potencial insercion en una base de datos noSQL como `mongoDB`, podemos servirnos de pandas para preparar los datos.

La estructura de datos de Python más similar a un documento NoSQL es el diccionario (dict).

¿Por qué un diccionario se parece a un documento NoSQL?

Los documentos en bases de datos NoSQL, como MongoDB, están estructurados en formato JSON o BSON, que tienen una representación muy similar a los diccionarios de Python:

- Clave-Valor: Al igual que en un diccionario de Python, los documentos en NoSQL almacenan datos en pares clave-valor.
- Estructura Anidada: Los valores en un diccionario pueden ser listas o incluso otros diccionarios, lo que permite modelar estructuras complejas, al igual que en NoSQL.
- Dinamismo: No es necesario definir un esquema fijo, lo que permite flexibilidad en la representación de datos.

MongoDB puede importar JSONLines fácilmente:

- Aunque MongoDB usa BSON internamente, su herramienta mongoimport permite cargar datos desde JSONLines (.jsonl).

JSONLines es más eficiente que JSON tradicional:

- Se puede leer línea por línea en streams.
- Se adapta mejor a cargas incrementales de datos en MongoDB.
- Podemos transformar fácilmente JSONLines en BSON:

Si necesitamos convertir un archivo JSONLines a BSON, MongoDB lo hace automáticamente al insertarlo.

¿Qué problemas pueden surgir al transformar un DataFrame en JSONLines?

Cuando conviertes un DataFrame a JSONLines (JSONL), pueden surgir varios problemas:

- Estructuras anidadas: Si hay listas o diccionarios en el DataFrame, pueden serializarse incorrectamente o como strings en lugar de objetos JSON.

- Valores faltantes (NaN, None): JSON no soporta NaN directamente, por lo que pueden convertirse en null, lo que podría no ser aceptado por ciertos sistemas. Es mejor reemplazarlos manualmente.

- Tipos de datos incompatibles: Columnas con fechas, categorías o números en formato NumPy pueden no exportarse correctamente. Es recomendable convertirlos a formatos estándar como strings antes de la conversión.

- Codificación de caracteres: Si hay caracteres especiales (acentos, emojis), es posible que se codifiquen incorrectamente en ciertos sistemas. Usar codificación UTF-8 ayuda a evitar problemas.

- Saltos de línea en los datos: Si hay textos con \n, podrían romper la estructura de JSONL al ser interpretados como nuevas líneas. Es mejor reemplazarlos con espaci

```python
# Añadir columnas con np.array y pd.Timestamp y transformarlas en JSONLines
# Ya sabemos que problemas nos puede dar esto así que:

# Convertiremos np.array en lista antes de exportar
# Convertiremos ps.Timestamp a string
```

```python
# Función para crear un heatmap de correlación
def plot_correlation_heatmap(df_filtered, title):
    plt.figure(figsize=(10, 8))
    sns.heatmap(df_filtered.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
    plt.title(title)
    plt.show()

# Heatmap para vinos tintos de alta calidad
plot_correlation_heatmap(red_high_quality, "Correlación - Vinos Tintos de Alta Calidad")

# Heatmap para vinos blancos de alta calidad
plot_correlation_heatmap(white_high_quality, "Correlación - Vinos Blancos de Alta Calidad")

# Heatmap para vinos tintos de baja calidad
plot_correlation_heatmap(red_low_quality, "Correlación - Vinos Tintos de Baja Calidad")

# Heatmap para vinos blancos de baja calidad
plot_correlation_heatmap(white_low_quality, "Correlación - Vinos Blancos de Baja Calidad")
```

Vemos que quality aparece en el heatmap de baja calidad pero no en el de alta calidad.

Esto no es un error, sino una característica de la correlación.

"quality" se excluye en alta calidad porque no cambia mucho (tiene valores constantes o casi constantes).

"quality" sí aparece en baja calidad porque hay más variabilidad.
